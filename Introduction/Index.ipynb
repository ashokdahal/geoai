# Course Overview

GeoAI (Geospatial Artificial Intelligence) brings together geospatial analysis and machine learning to uncover patterns, make predictions, and support data-driven decision-making in environmental and urban systems. This course introduces the core concepts and methods of AI and ML as applied to spatial data, aimed at students who already have experience with Python and geospatial data but are new to machine learning.

Over six sessions, students will build a strong foundation in supervised and unsupervised learning, understand how to engineer spatial features for ML, and explore both classical and deep learning methods for spatial prediction and image classification. The course emphasizes hands-on practice and real-world examples drawn from Earth observation, land use mapping, and environmental monitoring. The final session is dedicated to a small project where students apply GeoAI techniques to a dataset of their choice.

## Learning Outcomes

By the end of the course, students will be able to:

Understand the fundamentals of AI, Machine Learning, and Deep Learning in the context of geospatial problems.

Prepare and preprocess spatial datasets for use in ML workflows.

Train, validate, and interpret machine learning models (e.g., Decision Trees, Random Forests, SVMs).

Apply deep learning methods to remote sensing imagery using convolutional neural networks (CNNs).

Evaluate model performance and spatial accuracy, considering bias, overfitting, and uncertainty.

Develop and present a small GeoAI project addressing a real-world spatial question.

## Course Format

The course combines short lectures, live coding demonstrations, and practical exercises in Jupyter notebooks. Each session integrates theory and hands-on application, progressively building toward a final GeoAI project.

## Tools & Datasets

You will work with:

Platform: CRIB, Google Colab

Software: Python, Jupyter Notebook, GeoPandas, Rasterio, Scikit-learn, PyTorch.

Data sources: Sentinel-2 imagery, Copernicus Land Cover, SRTM elevation, and OpenStreetMap.

## Assessment

Students will complete practical exercises throughout the course and a final mini-project demonstrating their ability to apply GeoAI techniques to a geospatial challenge of their choice.
